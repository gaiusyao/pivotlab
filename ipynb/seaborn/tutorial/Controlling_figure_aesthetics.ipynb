```python
# -*-coding:utf-8 -*-
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(sum(map(ord, "aesthetics")))
```

```python
#定义一个含偏移的正弦图像
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
#调用sinplot, 使用matplotlib默认设置的图形效果
sinplot()
```

```python
#切换成seaborn, 再次调用sinplot, 查看seaborn设置的图形效果
#在seaborn 0.8版本以前, 是通过import切换, 而非set()
sns.set()
sinplot()
```

```python
#通过set_style()函数设置样式, 有5个seaborn主题, darkgrid为默认主题(见上图)
#下图为whitegrid主题示例
sns.set_style("whitegrid")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data);
```

```python
#dark主题
sns.set_style("dark")
sinplot()
```

```python
#white主题
sns.set_style("white")
sinplot()
```

```python
#ticks主题
sns.set_style("ticks")
sinplot()
```

```python
sinplot()
#通过despine()函数移除上方和右方坐标轴上不需要的边框
sns.despine()
```

```python
f, ax = plt.subplots()
sns.violinplot(data=data)
#offset设置两坐标轴离开距离, trim参数限制留存的边框范围
sns.despine(offset=10, trim=True); 
```

```python
sns.set_style("whitegrid")
sns.boxplot(data=data, palette="deep")
sns.despine(left=True) #通过往despine()中添加参数去除左边框
```

```python
#在with语句中套用axes_style()临时设置参数
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

```python
sns.axes_style() #调用axes_style()函数, 查看包含哪些参数
```

```python
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
sinplot()
```

```python
sns.set() #重置参数
sns.set_context("paper") #通过set_context()函数设置预设语境
#四种预设, 按相对尺寸的顺序(线条越来越粗) , 分别是paper, notebook(default), talk, and poster
sinplot()
```

```python
sns.set_context("talk")
sinplot()
```

```python
sns.set_context("poster")
sinplot()
```

```python
#通过更改context还可以独立地扩展字体元素的大小
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
sinplot()
```
