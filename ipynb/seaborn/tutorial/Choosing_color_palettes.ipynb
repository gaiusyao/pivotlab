```python
# -*-coding:utf-8 -*-
%matplotlib inline
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={"figure.figsize": (6, 6)})
np.random.seed(sum(map(ord, "palettes")))
```

```python
#通过color_palette()创建调色板
current_palette = sns.color_palette() #seaborn默认颜色主题
sns.palplot(current_palette) 
```

```python
#默认颜色主题共有六种不同的变化：deep, muted, pastel, bright, dark, and colorblind
current_palette = sns.color_palette("dark") #直接传入对应的参数即可变化
sns.palplot(current_palette)
```

```python
#当有六个以上的类别需要区分时, 最简单是是在一个圆形的颜色空间绘制均匀间隔的颜色
sns.palplot(sns.color_palette("hls", 8)) #最常用的是使用hls的颜色空间
```

```python
sns.palplot(sns.hls_palette(8, l=.3, s=.8)) #通过hls_palette()函数控制颜色的亮度和饱和度
```

```python
#seaborn提供了一个husl接口, 使得均匀间隔的色调很容易被选择，同时保持明显的亮度和饱和度更一致
sns.palplot(sns.color_palette("husl", 8))
```

```python
#另一种赏心悦目的分类调色板来自于Color Brewer
#它的优点之一, 是提供了一些关于色盲的无障碍阅读指导
sns.palplot(sns.color_palette("Paired"))
```

```python
#一般来说, 避免使用红色和绿色来表示颜色以区分元素是一个不错的主意
sns.palplot(sns.color_palette("Set2", 10))
```

```python
#设置一组自定义颜色
flatui = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
sns.palplot(sns.color_palette(flatui))
```

```python
#使用xkcd的命名颜色
plt.plot([0, 1], [0, 1], sns.xkcd_rgb["pale red"], lw=3)
plt.plot([0, 1], [0, 2], sns.xkcd_rgb["medium green"], lw=3)
plt.plot([0, 1], [0, 3], sns.xkcd_rgb["denim blue"], lw=3);
```

```python
colors = ["windows blue", "amber", "greyish", "faded green", "dusty purple"]
sns.palplot(sns.xkcd_palette(colors))
```

```python
#连续调色板
sns.palplot(sns.color_palette("Blues"))
```

```python
#添加_r后缀翻转渐变
sns.palplot(sns.color_palette("BuGn_r"))
```

```python
#添加_d后缀获得暗处理颜色
sns.palplot(sns.color_palette("GnBu_d"))
```

```python
#使用cubehelix_palette()函数创建连续调色板
#cubehelix调色板系统具有线性增加或降低亮度和色调变化顺序的调色板
sns.palplot(sns.color_palette("cubehelix", 8))
```

```python
sns.palplot(sns.cubehelix_palette(8))
```

```python
sns.palplot(sns.cubehelix_palette(8, start=.5, rot=-.75))
```

```python
#控制断点的亮度, 并对调结果顺序
sns.palplot(sns.cubehelix_palette(8, start=2, rot=0, dark=0, light=.95, reverse=True))
```

```python
#让调色板返回一个可以用as_cmap=True传入seaborn或matplotlib函数的颜色映射对象
x, y = np.random.multivariate_normal([0, 0], [[1, -.5], [-.5, 1]], size=300).T
cmap = sns.cubehelix_palette(light=1, as_cmap=True)
sns.kdeplot(x, y, cmap=cmap, shade=True);
```

```python
#使用light_palette() 和dark_palette()调用定制连续调色板
sns.palplot(sns.light_palette("green"))
sns.palplot(sns.dark_palette("purple"))
```

```python
#反转顺序
sns.palplot(sns.light_palette("navy", reverse=True))
```

```python
#创建一个颜色映射对象取代颜色列表
pal = sns.dark_palette("palegreen", as_cmap=True)
sns.kdeplot(x, y, cmap=pal);
```

```python
#默认情况下, 任何有效的matplotlib颜色可以传递给input参数
sns.palplot(sns.light_palette((210, 90, 60), input="husl"))
sns.palplot(sns.dark_palette("muted purple", input="xkcd"))
```

```python
#离散调色板用于可能无论大的低的值和大的高的值都非常重要的数据
#Color Brewer颜色字典里拥有一套精心挑选的离散颜色映射
sns.palplot(sns.color_palette("BrBG", 7))
sns.palplot(sns.color_palette("RdBu_r", 7))
```

```python
#coolwarm色板在中间值和极端之间并没有太大的对比
sns.palplot(sns.color_palette("coolwarm", 7))
```

```python
#使用diverging_palette()函数为离散的数据创建一个定制的颜色映射
sns.palplot(sns.diverging_palette(220, 20, n=7))
sns.palplot(sns.diverging_palette(145, 280, s=85, l=25, n=7))
```

```python
sns.palplot(sns.diverging_palette(10, 220, sep=80, n=7)) #sep参数控制中间区域的两个渐变的宽度
```

```python
#使用中间的色调来选择调色，而不是用亮度
sns.palplot(sns.diverging_palette(255, 133, l=60, n=7, center="dark"))
```

```python
#set_palette()函数更改默认的matplotlib参数，以便成为所有的调色板配置
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sns.set_palette("husl")
sinplot()
```

```python
#在with块中使用以达到临时更改调色板的目的
with sns.color_palette("PuBuGn_d"):
    sinplot()
```
