```python
#!/usr/bin/env python3.6
# -*-coding:utf-8 -*-
%matplotlib inline
import seaborn as sns
import numpy as np
import pandas as pd
from numpy import median

sns.set_style("whitegrid")
tips = sns.load_dataset("tips") # 载入自带数据集
# x轴为分类变量day,y轴为数值变量total_bill，利用颜色再对sex分类
antV = ["#61a5e8", "#7ecf51", "#eecb5f", "#9570e5", "#e3935d", "#e16757", "#605ff0"] # antV配色
with sns.color_palette(antV):
    ax = sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
```

```python
with sns.color_palette(antV):
    ax = sns.barplot(x="day", y="tip", data=tips, estimator=median)
#设置中位数为计算函数，注意y轴已显示为median
```

```python
titanic = sns.load_dataset("titanic")
with sns.color_palette(antV):
    ax = sns.countplot(x="class", hue="who", data=titanic)
```

```python
planets = sns.load_dataset("planets")
years = np.arange(2000, 2015)
# 选择调色板，绘图参数与顺序，factorplot是一种分类变量的集合作图方式
g = sns.factorplot(x="year", data=planets, kind="count", palette="BuPu", size=6, aspect=1.5, order=years)
g.set_xticklabels(step=2) # 设置x轴标签间距为2年
```
