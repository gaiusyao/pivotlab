```python
from pyecharts import Bar, Line
import pandas as pd
import numpy.random as random
```

```python
random.seed(111)

def CreateDataSet(Number=1):
    
    Output = []
    
    for i in range(Number):
        
        # 创建一个日期范围
        rng = pd.date_range(start='1/1/2015', end='12/31/2017')
        
        # 创建随机数据
        data = random.randint(low=25, high=1000, size=len(rng))
        
        # 状态池
        status = ['现货', '缺货', '下架', '预售', '暂不销售']
        
        # 随机列出一个状态列表
        random_status = [status[random.randint(low=0,high=len(status))] for i in range(len(rng))]
        
        # 促销类型池
        types = ['无促销', '普通促销', '超级促销']
        
        # 随机列出一个促销类型列表
        random_types = [types[random.randint(low=0,high=len(types))] for i in range(len(rng))]
    
        Output.extend(zip(rng, random_status, random_types, data))
        
    return Output
```

```python
# 折线图示例
attr = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
v1 = [5, 20, 36, 10, 10, 100]
v2 = [55, 60, 16, 20, 15, 80]
line = Line("折线图示例")
line.add("商家A", attr, v1, mark_point=["average"])
line.add("商家B", attr, v2, is_smooth=True, mark_line=["max", "average"])
line
```

```python
dataset = CreateDataSet(3)
df = pd.DataFrame(data=dataset, columns=['日期','状态','促销类型','销量（件）'])
df.info()
```
